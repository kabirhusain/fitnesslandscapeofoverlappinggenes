```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns

import pickle

sns.set_context("talk", rc={"font.size": 15, "axes.titlesize": 15, "axes.labelsize": 15})
sns.set_style("whitegrid", {"grid.color": ".9", "grid.linestyle": "--",
                             "axes.edgecolor": ".6", "xtick.bottom": True, "ytick.left": True})

colorTable = {}
colorTable['k'] = [0,0,0]
colorTable['g'] = [27/255,158/255,119/255]
colorTable['o'] = [217/255,95/255,2/255]
```

```python
# Load the shuffled data
pf1 = "PF00072"
pf2 = "PF00009"
overlap = 211

with open(f"Data/20260327_{pf1}_{pf2}_{overlap}_shuffled_permuted_codes_results.pkl", "rb") as f:
    shuffled_data = pickle.load(f)
```

```python
standard_z = shuffled_data["standard"]["z"]

shuffled_z = [thisdat["z"] for thisdat in shuffled_data["shuffled"]]
permuted_z = [thisdat["z"] for thisdat in shuffled_data["permuted"]]
```

```python
fig = plt.figure(figsize = (8,3))
plt.hist(shuffled_z, bins = np.linspace(-2,2,20), color = "k")
plt.hist(permuted_z, bins = np.linspace(-2,2,20), color = colorTable['o'])

# vertical line for the standard z
plt.axvline(standard_z, color = colorTable['k'], linestyle = "--", linewidth = 3)
# plt.savefig(f"Panels/20260328_{pf1}_{pf2}_{overlap}_Shuffled_Permuted_z_histogram.svg", dpi = 300, bbox_inches = "tight")
```
